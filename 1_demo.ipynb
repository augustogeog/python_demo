## Tipos Compostos


* str - conjunto imutvel de caracteres delimitado por aspas simples ou duplas, cujos elementos so acessveis por indexao posicional 
* list - conjunto mutvel de quaisquer elementos, delimitado por colchetes, cujos elementos so separados por vrgulas e acessveis por indexao posicional
* tuple - conjunto imutvel delimitado por parntesis, cujos elementos so separados por vrgulas e acessveis por indexao posicional
* dict - conjunto de pares key:value, separados por vrgulas e delimitados por chaves
* set - conjunto de valores no duplicveis, separados por vrgulas e delimitados por chaves

````python
# declarar uma varivel texto que recebe uma string com os <caracteres aqui  uma string>

texto = "aqui  uma string"
````

````python
# declarar uma lista e atribu-la a uma varivel chamada lista_compras
# inserir produtos de mercado nessa lista

lista_compras = ['mamo', 'limo', 'caf', 'manteiga', 'coco', 'cerveja', 'carne'] 
````

````python
# chamar a varivel lista_compras

lista_compras
````

````python
# Com o construtor de tuplas, declarar uma tupla a partir da list lista_compras.
# Atribuir essa tupla  varivel tupla_compras

tupla_compras = tuple(lista_compras)
````

````python
# chamar a varivel tupla_compras

tupla_compras
````

````python
# faa um conjunto com os seguintes valores janeiro, janeiro, fevereiro, fevereiro, maro, maro, maro
# atribua o conjunto  varivel set_primeiro_trimestre

set_primeiro_trimestre = set(['janeiro', 'janeiro', 'fevereiro', 'fevereiro', 'maro', 'maro', 'maro'])
````

````python
# apresente na tela o conjunto set_primeiro_trimestre

print(set_primeiro_trimestre)
````

````python
# criar um dicinrio que contenha os trs objetos criados anteriorment (strint texto, lista lista_compras e tupla tupla_compras)
# atribuir  varivel dicionario

dicionario = {'texto':texto, 'lista':lista_compras, 'tup':tupla_compras, 'set': set_primeiro_trimestre}
````

````python
# apresente na tela o objeto da varivel dicionrio
print(dicionario)
````

````python
# faa o fatiamentos/slices da tupla na varivel tuplas

tupla_compras[:]
````
